```python
exp_100 = np.random.exponential(1, size=100)
exp_1000 = np.random.exponential(1, size=1000)
norm_100 = np.random.normal(0, 1, size=100)
norm_1000 = np.random.normal(0, 1, size=1000)
```

```python
print('Mean for exp_100:', np.mean(exp_100))
print('Mean for exp_1000:', np.mean(exp_1000))
print('Mean for norm_100:', np.mean(norm_100))
print('Mean for norm_1000:', np.mean(norm_1000))
```

```python
# Функция для вычисления и вывода результатов
def summarize(data, theoretical_mean, theoretical_var, name):
    sample_mean = np.mean(data)
    sample_var = np.var(data, ddof=1)  # несмещённая дисперсия
    print(f"{name}:")
    print(f"  Выборочное среднее: {sample_mean:.4f}, Теоретическое: {theoretical_mean}")
    print(f"  Дисперсия: {sample_var:.4f}, Теоретическая: {theoretical_var}\n")

# Экспоненциальное распределение: lambda=1, E[X]=1/λ=1, Var[X]=1/λ^2=1
summarize(exp_100, theoretical_mean=1, theoretical_var=1, name="Exponential 100")
summarize(exp_1000, theoretical_mean=1, theoretical_var=1, name="Exponential 1000")

# Нормальное распределение: mean=0, sigma^2=1
summarize(norm_100, theoretical_mean=0, theoretical_var=1, name="Normal 100")
summarize(norm_1000, theoretical_mean=0, theoretical_var=1, name="Normal 1000")
```

```python
def distance(x1,y1,x2,y2):
    return np.sqrt((x2-x1)**2+(y2-y1)**2)
def distance_line(number):
    a=10
    b=30
    distances=[]
    for i in range(number):
        x1=np.random.uniform(0,a)
        y1=np.random.uniform(0,b)
        
        x2=np.random.uniform(0,a)
        y2=np.random.uniform(0,b)

        distances.append(distance(x1,y1,x2,y2))
    distances.sort()
    return distances
# Выборки разного размера
d100 = distance_line(100)
d1000 = distance_line(1000)
d10000 = distance_line(10000)

# Средние значения
print("Среднее (100):", np.mean(d100))
print("Среднее (1000):", np.mean(d1000))
print("Среднее (10000):", np.mean(d10000))

# ---------- CDF ----------
plt.figure(dpi=100)
plt.title("Функция распределения расстояний (CDF)")
for data, label in zip([d100, d1000, d10000], ["100", "1000", "10000"]):
    sorted_data = np.sort(data)
    y = np.arange(1, len(sorted_data)+1) / len(sorted_data)
    plt.plot(sorted_data, y, label=f"n={label}")
plt.xlabel("Расстояние")
plt.ylabel("F(x)")
plt.legend()
plt.show()

# ---------- PDF ----------
plt.figure(dpi=100)
plt.title("Плотность распределения расстояний (PDF)")
plt.hist(d100, bins=30, density=True, alpha=0.5, label="100")
plt.hist(d1000, bins=30, density=True, alpha=0.5, label="1000")
plt.hist(d10000, bins=30, density=True, alpha=0.5, label="10000")
plt.xlabel("Расстояние")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
