```python
numbers = np.random.poisson(5,1000)
mean = np.mean(numbers) # Вычисление математического ожидания (среднего)
variance = np.var(numbers) # Вычисление дисперсии
print(f"Математическое ожидание: {mean}")
print(f"Дисперсия: {variance}")
plt.hist(numbers, bins=range(min(numbers), max(numbers)+2), alpha=0.7, color='skyblue', edgecolor='black', align='left')
plt.title(f"Гистограмма распределения Пуассона (lambda={5})")
plt.xlabel("Значение")
plt.ylabel("Частота")
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
a = 10 # Длина стороны квадрата
x = np.random.uniform(0, a) # Генерация случайной точки (x, y) в квадрате
y = np.random.uniform(0, a)
print(f"Случайная точка: ({x:.3f}, {y:.3f})")
plt.figure(figsize=(5,5))
plt.plot(x, y, 'ro', markersize=8, label='Случайная точка')
plt.xlim(0, a)
plt.ylim(0, a)
plt.gca().set_aspect('equal', adjustable='box')
plt.title(f'Случайная точка в квадрате со стороной {a}')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()
plt.show()
```
